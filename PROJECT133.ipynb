```python
import csv
import plotly.express as px

row = []
with open ('/content/PRO-Stars-Dataset-CSVs/star_with_gravity.csv') as f:
  csvreader = csv.reader(f)
  for i in csvreader:
    row.append(i)
headers = row[0]
planet_data_rows = row[1:]
```

```python
planet_radius = []
planet_mass = []
for i in planet_data_rows:
  planet_radius.append(i[4])
  planet_mass.append(i[3])
graph = px.scatter(x=planet_mass, y=planet_radius)
graph.show()
```
